#### Query Data

```python
df_only_huskies = df[
df['fran_id'] == 'Huskies'
]
```

```python
df_only_huskies.head()
```

```python
#conditional filter with bracket and &
df[
(df['fran_id'] == 'Huskies') &
(df['pts'] > 50) &
(df['opp_pts'] > 80)
]
```

```python
df[
    (df['fran_id'] == 'Lakers') &
    (df['year_id'] == 1992) &
    (df['notes'].notnull())
]
```

```python
df.query('fran_id == "Lakers" & year_id == 1992 & notes.notnull')
```
